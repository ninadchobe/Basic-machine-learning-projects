```python
data=data.drop('Publisher',axis=1)
```

```python
data=data.drop('Name',axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder
data['Genre']=data['Genre'].astype('category')
data['Platform']=data['Platform'].astype('category')
#Assigning numerical values and storing it in another columns
data['Gen_new']=data['Genre'].cat.codes
data['plat_new']=data['Platform'].cat.codes

enc=OneHotEncoder()

enc_data=pd.DataFrame(enc.fit_transform(data[['Gen_new','plat_new']]).toarray())
new_data=data.join(enc_data)
```

```python
new_data=new_data.drop('Genre',axis=1)
new_data=new_data.drop('Platform',axis=1)
```

```python
new_data.fillna(0,inplace=True)
```

```python
X=new_data.drop('Global_Sales',axis=1)
Y=new_data['Global_Sales']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train,X_test,Y_train,Y_test=tts(X,Y,test_size=0.25)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
print(X_test)
```

```python
lrg=LinearRegression()
model=lrg.fit(X_train,Y_train)
Y_pred=lrg.predict(X_test)
```

```python
import math
```

```python
MSE = np.square(np.subtract(Y_test,Y_pred)).mean() 
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```
